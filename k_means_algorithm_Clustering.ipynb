```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('student_clustering.csv')
df
```

```python
df.info()
```

```python
sns.scatterplot(data=df,x='cgpa',y='iq')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)# number of centroids will be 2
```

```python
model.fit_predict(df)
```

```python
model.inertia_ #points nearest to centroids
```

```python
inr_list = [] #list of inertia of centroid
```

```python
for i in range(1,15): # loop to find number of clusters on the basis of inertia

  model = KMeans(n_clusters = i)
  model.fit_predict(df)
  inr = model.inertia_
  inr_list.append(inr)
```

```python
print(inr_list)
```

```python
plt.plot(range(1,15), inr_list)

plt.xlabel("Value of K")
plt.ylabel("Inertia")
plt.title("K vs. Inertia")

plt.show()
```
