```python
#encoding-normal and ordinal
#Nominal data= one-hot encoding  data preprocessing technique used to convert categorical data into a binary numerical format (0s and 1s)
#ordinal data= ordinal encoding   a data preprocessing technique used in machine learning to convert categorical data into numerical values by assigning a unique, ordered integer to each category
df=pd.read_csv('titanic.csv')
df
```

```python
df= pd.get_dummies(df,columns=['Sex'],drop_first= True)#forms dummy column of one orignal column
print(df)
```

```python
#feature scaling and normalizaation -data preprocessing technique that normalizes the range of independent variables or features

#FEATURE
X=df[['Age','Pclass','Sex_male']]
#TARGET
y=df['Survived']

```
