```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

df=pd.read_csv('titanic.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns='Cabin')
```

```python
df['Age'].skew()
```

```python
avg=df['Age'].mean()
df['Age']=df['Age'].fillna(avg)
df.info()
```

```python
m=df['Embarked'].mode()[0]
df['Embaeked']=df['Embarked'].fillna(m)
df.info()
```

```python
df.to_csv('titanic_cleaned.csv',index=False)
files.download('titanic_cleaned.csv')
```
