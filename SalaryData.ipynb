```python
import numpy as np
import pandas as pd
df=pd.read_csv('Salary_Data.csv')
df.info()
```

```python
df['Education Level'].unique()
```

```python
df['Education Level']=df['Education Level'].replace("Bachelor's","Bachelor's Degree") #.unique() method in Pandas extracts all distinct values from a DataFrame column (Series).
df['Education Level'].unique()
```

```python
df['Education Level']=df['Education Level'].replace("Master's","Master's Degree")
df['Education Level'].unique()
```

```python
df['Education Level']=df['Education Level'].replace("phD","PhD")
df['Education Level'].unique()
```

```python
df['Education Level']=df['Education Level'].fillna(df['Education Level'].mode()[0])
```

```python
df['Years of Experience'].skew()
```

```python
df['Years of Experience']=df['Years of Experience'].fillna(df['Years of Experience'].mean())
```

```python
df['Salary'].skew()
```

```python
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
```

```python
df.info()
```

```python
df= pd.get_dummies(df,columns=['Gender'],drop_first= True)#forms dummy column of one orignal column
print(df)
```

ORDINAL ENCODING - a data preprocessing technique that converts categorical text data into numerical values by assigning a unique, ordered integer to each category.

```python
from sklearn.preprocessing import OrdinalEncoder

oe= OrdinalEncoder(categories=[["High School","Bachelor's Degree","Master's Degree","PhD"]])#object creation

df['Education Level']=oe.fit_transform(df[['Education Level']]) #find and transform

df.head()
```
